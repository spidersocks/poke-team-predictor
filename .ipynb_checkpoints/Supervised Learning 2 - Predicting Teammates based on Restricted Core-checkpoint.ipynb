```python
#Let's ensure we got all 30 restricteds properly named
# Get all columns that contain 'species' in the name
species_columns = [col for col in battle_df.columns if 'species' in col.lower()]

# Flatten all values into one Series
all_species = pd.Series(battle_df[species_columns].values.ravel())

# Define the restricted set
restricted_set = set(restricted_pokemon_reg_I)

# Filter to only restricted Pokémon and get unique ones
restricted_present = set(all_species[all_species.isin(restricted_set)])

# Output the result
print(f"Number of unique restricted Pokémon in battle_df: {len(restricted_present)}")
print("Restricted Pokémon present:")
print(sorted(restricted_present))
#Based on the prints, looks good!
```

Map each team into core and teammates

```python
#this step should take 50s - 1 minute

# Step 1: Melt to long format
species_cols = [f'{prefix}_species_{i}' for prefix in ['p1', 'p2'] for i in range(1, 7)]
long_df = battle_df[species_cols].copy()

# Add battle + team IDs
long_df['battle_id'] = long_df.index
long_df = long_df.melt(id_vars='battle_id', var_name='slot', value_name='pokemon')
long_df['team'] = long_df['slot'].str.extract(r'(p\d)').iloc[:, 0]
long_df['slot_num'] = long_df['slot'].str.extract(r'(\d)$').astype(int)

# Step 2: group by battle_id and team
grouped = long_df.dropna().groupby(['battle_id', 'team'])['pokemon'].apply(list).reset_index()

# Step 3: classify into core + teammates and filter
def classify_team(pokemon_list):
    core = sorted([p for p in pokemon_list if p in restricted_set])
    teammates = sorted([p for p in pokemon_list if p not in restricted_set])
    if len(core) == 2 and len(teammates) == 4:
        return pd.Series({'core': tuple(core), 'teammates': tuple(teammates)})
    return pd.Series({'core': None, 'teammates': None})

core_team_df = grouped['pokemon'].apply(classify_team)
core_team_df = pd.concat([grouped[['battle_id', 'team']], core_team_df], axis=1)
core_team_df = core_team_df.dropna()
```

# 3. Feature Engineering
- One-hot encode restricted cores into `X_df`
- One-hot encode teammates into `Y_df`

One-hot encode or embed the 2-pokemon core

```python
# Extract all unique restricted Pokémon from the core column
unique_restricted = sorted({mon for core in core_team_df['core'] for mon in core})

# One-hot encode the 2-Pokémon core into a binary vector for each team
def one_hot_core(core_pair):
    return [int(mon in core_pair) for mon in unique_restricted]

# Create the feature matrix
X = core_team_df['core'].apply(one_hot_core).tolist()
X_df = pd.DataFrame(X, columns=[f"core_{mon}" for mon in unique_restricted])
X_df
```

```python
#this cell should take about 1 minute
# --- General One-Hot Encoding Function ---
def one_hot_encode(column, unique_values):
    return column.apply(lambda entry: [int(mon in entry) for mon in unique_values]).tolist()

# Extract all unique Pokémon from cores and teammates
unique_restricted = sorted({mon for core in core_team_df['core'] for mon in core})
unique_teammates = sorted({mon for team in core_team_df['teammates'] for mon in team})

# Apply the general encoder
X = one_hot_encode(core_team_df['core'], unique_restricted)
Y = one_hot_encode(core_team_df['teammates'], unique_teammates)

# Wrap in DataFrames
X_df = pd.DataFrame(X, columns=[f"core_{mon}" for mon in unique_restricted])


Y_df = pd.DataFrame(Y, columns=[f"teammate_{mon}" for mon in unique_teammates])
Y_df_pokemon = Y_df.copy()   # Save the original, Pokémon-level one-hot dataframe (This is for our function at the bottom of the notebook)

X_df
```

# 4. Recommended: Teammate Clustering with K-Means
- Map teammates to cluster labels
- Convert `Y_df` to cluster-based multi-label format

BEFORE- was doing clustering based on stats. NOW- doing co-occurance on teams.

1. Build a Co-Occurrence Matrix

```python
cooccur_dict = defaultdict(set)

# Scan all rows
for _, row in battle_df.iterrows():
    # Get all 6 mon names from player 1
    mons = [row[f"p1_species_{i}"] for i in range(1, 7)]
    mons = [mon for mon in mons if pd.notna(mon)]

    # Separate into restricteds and teammates
    restricteds = [mon for mon in mons if mon in restricted_set]
    teammates = [mon for mon in mons if mon not in restricted_set]

    # If we have exactly 2 restricteds, record them as the core
    if len(restricteds) == 2:
        core = tuple(sorted(restricteds))
        for mon in teammates:
            cooccur_dict[mon].add(core)
```

```python
# Step 2: Create a sorted list of all unique cores observed
all_cores = sorted({core for cores in cooccur_dict.values() for core in cores})

# Convert each core from a tuple to a readable string (for column labels)
core_labels = [f"{a}+{b}" for (a, b) in all_cores]

# Initialize data and index lists
data = []
index = []

# Build the binary matrix row by row
for mon, cores in cooccur_dict.items():
    row = [1 if core in cores else 0 for core in all_cores]
    data.append(row)
    index.append(mon)

# Convert to DataFrame: Pokémon (rows) x Restricted Cores (columns)
cooccur_df = pd.DataFrame(data, index=index, columns=core_labels)
cooccur_df
```

2. Cluster Pokémon by Co-Occurrence

```python
# Evaluate inertia for k from 1 to 10
inertias = []
ks = range(1, 11)

for k in ks:
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(cooccur_df)  # Use co-occurrence matrix here
    inertias.append(km.inertia_)

# Plot the elbow
plt.figure()
plt.plot(ks, inertias, marker='o')
plt.title("Elbow Method for Optimal k (Co-Occurrence Clustering)")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Inertia")
plt.xticks(ks)
plt.grid(True)
plt.show()
```

```python
# Step 2: Run K-Means on cooccurrence matrix
k = 5  # optimal k from elbow plot
kmeans = KMeans(n_clusters=k, random_state=42)
cluster_labels = kmeans.fit_predict(cooccur_df)

# Step 3: Map each Pokémon to its cluster
teammate_to_cluster = dict(zip(cooccur_df.index, cluster_labels))

# Step 4: Convert Y_df to cluster-based multi-label format
cluster_y = np.zeros((len(Y_df), k), dtype=int)

for i, row in Y_df.iterrows():
    for col in row[row == 1].index:
        mon = col.replace("teammate_", "")
        if mon in teammate_to_cluster:
            cluster_idx = teammate_to_cluster[mon]
            cluster_y[i, cluster_idx] = 1

# Replace Y_df with cluster-based labels
Y_df = pd.DataFrame(cluster_y, columns=[f"cluster_{i}" for i in range(k)])
Y_df
```

# 5. Modeling
- Train/test split
- Model definition (e.g., RandomForest, LogisticRegression)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
#Y_df_pokemon is very large, so this portion should take about 10 minutes
# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X_df, Y_df_pokemon, test_size=0.2, random_state=42)

#model = OneVsRestClassifier(LogisticRegression(max_iter=500, solver='saga', penalty='l1')) #using l1 lasso regression, because most features are just useless noise
# Train a multi-output random forest
model = MultiOutputClassifier(RandomForestClassifier(n_estimators=50, random_state=42))
model.fit(X_train, Y_train)
```

# 6. Evaluation
- Hamming Loss
- Exact Match Accuracy
- Per-label F1 scores

```python
# Predict
Y_pred = model.predict(X_test)

# Convert predictions to DataFrame
Y_pred_df = pd.DataFrame(Y_pred, columns=Y_test.columns, index=Y_test.index)

# Evaluate
hamming = hamming_loss(Y_test, Y_pred_df)
exact_match = (Y_test.values == Y_pred_df.values).all(axis=1).mean()
report = classification_report(Y_test, Y_pred_df, zero_division=0)

print(f'## With K={k} and N={N}')
print(f"Hamming Loss: {hamming:.4f}")
print(f"Exact Match Accuracy: {exact_match:.4f}")
print("\nClassification Report (partial):")
print(report[:1000])
```

# TURN INTO A FUNCTION

We are going to run a co-occurrence matrix prediction alongside a model prediction to compare results and see if we uncover nuance.

```python
def recommend_teammates_from_cooccur(core, top_n=10):
    core = tuple(sorted(core))
    # Count how many times each mon co-occurred with this core
    teammate_counter = Counter()

    for mon, core_set in cooccur_dict.items():
        if core in core_set:
            teammate_counter[mon] += 1

    total = sum(teammate_counter.values())
    if total == 0:
        return f"No data available for core: {core}"

    results = [(mon, round(count / total, 3)) for mon, count in teammate_counter.most_common(top_n)]
    return pd.DataFrame(results, columns=["Teammate", "Probability"])

recommend_teammates_from_cooccur(("Calyrex-Shadow", "Zamazenta-Crowned"))
```

```python
def predict_teammates(core, model=model, X=X_df, Y=Y_df_pokemon, top_n=10):
    core = list(core)
    input_row = pd.DataFrame([0]*X.shape[1], index=X.columns).T

    for mon in core:
        col = f"core_{mon}"
        if col in input_row.columns:
            input_row.at[0, col] = 1
        else:
            print(f"Warning: {col} not in input features")

    probs = []
    for prob_arr in model.predict_proba(input_row):
        # prob_arr shape is (1, n_classes)
        if prob_arr.shape[1] == 2:
            probs.append(prob_arr[0, 1])  # Probability for class 1
        else:
            # Only one class ever appeared in train data; which class?
            # model.classes_ for that estimator tells us.
            # If the only class is 1 (positive), prob is always 1.0; if only 0, it's 0.0
            clf = prob_arr
            label_idx = model.estimators_[len(probs)].classes_[0]
            probs.append(1.0 if label_idx == 1 else 0.0)
    teammate_names = [col.replace("teammate_", "") for col in Y.columns]

    results = sorted(zip(teammate_names, probs), key=lambda x: x[1], reverse=True)[:top_n]
    return pd.DataFrame(results, columns=["Teammate", "Predicted Probability"])
predict_teammates(("Calyrex-Shadow", "Zamazenta-Crowned"))
```

```python
# core_team_df['core'] is a list/tuple of restricted mons for each team
core_counts = Counter(tuple(sorted(core)) for core in core_team_df['core'] if len(core) == 2)

top_n = 4
most_common_cores = core_counts.most_common(top_n)

print("Top {top_n} most common restricted cores:")
for core, count in most_common_cores:
    print(f"{core}: {count} teams")
```
